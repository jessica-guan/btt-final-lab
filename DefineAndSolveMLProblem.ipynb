```python
# replace missing values with means
df.fillna(df.mean(numeric_only = True), inplace = True)
```

```python
# peform winsorization
import scipy.stats as stats
for column in df.select_dtypes(include = ['number']).columns:
    df[column] = stats.mstats.winsorize(df[column], limits = [0.01, 0.01])
```

```python
# assign social support as the feature for predicting life ladder as they appeared to have a relatively strong correlation
X = df['Social Support'].to_frame()
y = df['Life Ladder']
```

```python
# split data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# initialize model
model = LinearRegression()
```

```python
# fit model to training data
model.fit(X_train, y_train)
```

```python
# visualize actual vs predicted results
y_pred = model.predict(X_test)
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('True Values vs. Predicted Values')
plt.show()
```

```python
# evaluate model
print("Mean Absolute Error Score:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error Score:", mean_squared_error(y_test, y_pred))
print("R-squared Score:", r2_score(y_test, y_pred))
```

```python
# 3-fold cross validation
scores = cross_val_score(model, X, y, cv = 3, scoring = 'r2')
print("Cross-validation R-squared scores:", scores)
print("Mean R-squared score:", scores.mean())
```
